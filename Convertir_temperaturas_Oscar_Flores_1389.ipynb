## 1. Fórmulas de Conversión

Para realizar estas conversiones, utilizamos ecuaciones lineales basadas en los puntos de congelación y ebullición del agua.

* **De Celsius a Fahrenheit:**
    $$°F = (°C \times \frac{9}{5}) + 32$$
    *(Multiplicas por 1.8 y sumas 32)*

* **De Fahrenheit a Celsius:**
    $$°C = \frac{(°F - 32) \times 5}{9}$$
    *(Restas 32 y multiplicas por 0.555...)*

## 2. Ejemplos Resueltos Paso a Paso

### Ejemplo 1: Convertir 25°C a Fahrenheit (Temperatura ambiente típica)
1.  **Identificar la fórmula:** $°F = (°C \times 1.8) + 32$
2.  **Sustituir el valor:** $°F = (25 \times 1.8) + 32$
3.  **Multiplicar:** $25 \times 1.8 = 45$
4.  **Sumar 32:** $45 + 32 = 77$
**Resultado:** 25°C equivalen a **77°F**.

### Ejemplo 2: Convertir 98.6°F a Celsius (Temperatura corporal promedio)
1.  **Identificar la fórmula:** $°C = (98.6 - 32) \times \frac{5}{9}$
2.  **Restar 32:** $98.6 - 32 = 66.6$
3.  **Multiplicar por 5:** $66.6 \times 5 = 333$
4.  **Dividir entre 9:** $333 / 9 = 37$
**Resultado:** 98.6°F equivalen a **37°C**.

## 3. Código en Python

Aquí tienes el programa interactivo siguiendo todos tus requerimientos técnicos.

```python
def convertirCF():
    celsius = float(input("\nIngresa los grados Celsius a convertir: "))
    fahrenheit = (celsius * 9/5) + 32
    print(f"--- {celsius}°C es igual a {fahrenheit:.2f}°F ---")

def convertirFC():
    fahrenheit = float(input("\nIngresa los grados Fahrenheit a convertir: "))
    celsius = (fahrenheit - 32) * 5/9
    print(f"--- {fahrenheit}°F es igual a {celsius:.2f}°C ---")

def menu():
    while True:
        print("\n=== CONVERSOR DE TEMPERATURA ===")
        print("1. Convertir grados Celsius a Fahrenheit")
        print("2. Convertir grados Fahrenheit a Celsius")
        print("3. Salir")
        
        opcion = input("\nSelecciona una opción (1-3): ")

        if opcion == '1':
            convertirCF()
        elif opcion == '2':
            convertirFC()
        elif opcion == '3':
            print("Saliendo del programa... ¡Hasta luego!")
            break
        else:
            print("Opción no válida, por favor intenta de nuevo.")

# Llamada inicial al menú
if __name__ == "__main__":
    menu()
```

### Explicación del código:
* **`def`**: Define las funciones para mantener el código organizado y reutilizable.
* **`float()`**: Permite que el usuario ingrese números con decimales.
* **`while True`**: Crea un bucle infinito que solo se detiene cuando el usuario elige la opción de salida.
* **`f-strings`**: Se usan en los `print` para mostrar los resultados de forma limpia, limitando a 2 decimales con `:.2f`.
* **`break`**: Rompe el ciclo del menú inmediatamente cuando se selecciona la opción 3.

````python
def convertirCF():
    celsius = float(input("\nIngresa los grados Celsius a convertir: "))
    fahrenheit = (celsius * 9/5) + 32
    print(f"--- {celsius}°C es igual a {fahrenheit:.2f}°F ---")

def convertirFC():
    fahrenheit = float(input("\nIngresa los grados Fahrenheit a convertir: "))
    celsius = (fahrenheit - 32) * 5/9
    print(f"--- {fahrenheit}°F es igual a {celsius:.2f}°C ---")

def menu():
    while True:
        print("\n=== CONVERSOR DE TEMPERATURA ===")
        print("1. Convertir grados Celsius a Fahrenheit")
        print("2. Convertir grados Fahrenheit a Celsius")
        print("3. Salir")

        opcion = input("\nSelecciona una opción (1-3): ")

        if opcion == '1':
            convertirCF()
        elif opcion == '2':
            convertirFC()
        elif opcion == '3':
            print("Saliendo del programa... ¡Hasta luego!")
            print("¡Oscar Flores NL = 19!")
            break
        else:
            print("Opción no válida, por favor intenta de nuevo.")

# Llamada inicial al menú
if __name__ == "__main__":
    menu()
````
